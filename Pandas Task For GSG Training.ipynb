# Step 2

### Preprocessing data; deleting unnecessary columns and handling missing values

```python
# let's check the columns that contains null values   ====>   isnull()  ==> True

df.isnull().sum()
    
  
```

As we can see, the following columns contain null values: 101 values missing in the 'Certificate' column, 157 values missing in the 'Meta_score' column & 169 values missing in the 'Gross' column.

now we need to handle these values:

```python
# We can fill null values with mean value for Meta_score & Gross Columns:

df['Meta_score'].fillna(value=df['Meta_score'].mean(), inplace=True)
df.head(80)
```

```python
df.isnull().sum()
```

```python
df['Gross'].isnull()
```

```python
#in order to be able to convert the formart of 'ddd,ddd,ddd' into float or integer we need to get rid of the ',' between numbers:

df['Gross'] = df['Gross'].str.replace(",","")
df['Gross'] = df['Gross'].astype(float)

df['Gross'] = df['Gross'].fillna(value = df['Gross'].mean())
```

```python
df['Gross']
```

```python
# to make sure we have no missing value in the 'Gross' Series,
df['Gross'].isnull().sum()
```

Now we only have the 'certificate' Column having some missing values:

We can get rid of the entire column as it seems to have no effect on the final rating.

```python
#removing the certificate column

df.drop(columns = 'Certificate',inplace = True)
df
```

# Step 3&4

```python
genres = df['Genre'].str.split(',').sum()
genres_list = [x.strip() for x in genres ]  # one list containing all genres avaliable on the df

genre_count = {}   #dictionary to save the counts of each genre
def count_genre(x):
    for i in x:
        genre_count[i] = genre_count.get(i, 0) + 1   #make dictionary for counting values
    return genre_count

count_genre(genres_list)
```

```python
d = {'Genre': genre_count.keys(), 'Count': genre_count.values()}
```

```python
clmn = len(genre_count)
clmn
```

```python
df2 = pd.DataFrame(data = d, index=[x for x in range(clmn)])
df2
```

```python
filt = (df2['Count'] == df2['Count'].max())
df2[filt]
```

Here we can see that the Drama genre is the most common genre among movies/TV shows with 724 occurance. 

Also, we can see the number of movies/TV shows associated with each genre.

### Step 6

```python
import numpy as np
```

```python
#Director_DF = df.groupby(['Director', 'Series_Title']).agg({'Series_Title': 'count','IMDB_Rating':'max'})
Director_DF = df.groupby(['Director', 'Series_Title', 'IMDB_Rating'], as_index=[True, False, False]).agg({'Series_Title': 'count'})
Director_DF.rename(columns={'Series_Title': 'Series_count'}, inplace=True)
Director_DF.head(20)
```

```python
Director_DF.sort_values(by='IMDB_Rating', ascending=False, inplace=True)
```

```python
Director_DF.head(100)  #sorted by max IMDB Rating
```

Here we can see data ordered as per the Director and then it is sorted according to its IMDB Rating. 

```python
#We can also get the number of series/TV shows associated with each Director and his max IMDB rating by:
Director_DF2 = df.groupby(['Director']).agg({'Series_Title': 'count','IMDB_Rating':'max'})
Director_DF2.rename(columns={'Series_Title': 'Series_count', 'IMDB_Rating':'Max_IMDB_Rating'}, inplace=True)
Director_DF2.sort_values('Max_IMDB_Rating', ascending=False, inplace=True)
Director_DF2.head(10)
```

We can see that 'Frank Darabont' got the highest IMDB rating for his movie: 'The Shawshank Redemption'.
